# Extracting data from json files

```python
#Required libraries for the program
import pandas as pd
import json
import os
```

## AGG

### Aggrate transation

```python
Agg_tran_path="pulse/data/aggregated/transaction/country/india/state"
Agg_state_tran_list=os.listdir(Agg_tran_path)
# Agg_state_tran_list

atd = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

for i in Agg_state_tran_list:
    p_i = os.path.join(Agg_tran_path, i, "")
    try:
        Agg_yr = os.listdir(p_i)
    except FileNotFoundError:
        print(f"Directory not found: {p_i}")
        continue

    for j in Agg_yr:
        p_j = os.path.join(p_i, j, "")
        try:
            Agg_yr_list = os.listdir(p_j)
        except FileNotFoundError:
            print(f"Directory not found: {p_j}")
            continue

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            # Check if the file is a JSON file
            if p_k.endswith('.json'):
                try:
                    with open(p_k, 'r') as Data:
                        D = json.load(Data)
                        for z in D['data']['transactionData']:
                            Name = z['name']
                            count = z['paymentInstruments'][0]['count']
                            amount = z['paymentInstruments'][0]['amount']
                            atd['Transaction_type'].append(Name)
                            atd['Transaction_count'].append(count)
                            atd['Transaction_amount'].append(amount)
                            atd['State'].append(i)
                            atd['Year'].append(j)
                            atd['Quarter'].append(int(k.strip('.json')))
                except FileNotFoundError:
                    print(f"File not found: {p_k}")
                    continue

Agg_Trans_df = pd.DataFrame(atd)


# Specify the directory path where you want to save the CSV file
output_dir = "extracted_data"

# Create the directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# File path for saving the CSV file inside the "data" folder
output_file_path = os.path.join(output_dir, "agg_transaction.csv")

# Save the DataFrame to the specified file path
Agg_Trans_df.to_csv(output_file_path, index=False)

# Display the DataFrame
Agg_Trans_df
```

### Aggrate user

```python
#This is to direct the path to get the data as states
Agg_user_path = "pulse/data/aggregated/user/country/india/state"
Agg_state_user_list = os.listdir(Agg_user_path)
# Agg_state_user_list


Agg_user = {'State': [], 'Year': [], 'Quater': [], 'brands': [], 'Count': [], 'Percentage': []}

for i in Agg_state_user_list:
    p_i = os.path.join(Agg_user_path, i)  # Use os.path.join to construct paths
    Agg_yr = os.listdir(p_i)
    
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Use os.path.join to construct paths
        Agg_yr_list = os.listdir(p_j)
        
        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Use os.path.join to construct paths
            
            try:
                with open(p_k, 'r') as Data:
                    B = json.load(Data)
                    
                    for w in B["data"]["usersByDevice"]:
                        brand_name = w["brand"]
                        count_ = w["count"]
                        ALL_percentage = w["percentage"]
                        
                        Agg_user["brands"].append(brand_name)
                        Agg_user["Count"].append(count_)
                        Agg_user["Percentage"].append(ALL_percentage)
                        Agg_user["State"].append(i)
                        Agg_user["Year"].append(j)
                        Agg_user["Quater"].append(int(k.strip('.json')))
            except:
                continue

# Create a DataFrame
Agg_user_df = pd.DataFrame(Agg_user)

# Specify the directory path where you want to save the CSV file
output_dir = "extracted_data"

# File path for saving the CSV file inside the "data" folder
output_file_path = os.path.join(output_dir, "agg_user.csv")

# Save the DataFrame to the specified file path
Agg_user_df.to_csv(output_file_path, index=False)

# Display the DataFrame
Agg_user_df
```

## MAP

### MAP TRANSACTION

```python
map_tran_path="pulse/data/map/transaction/hover/country/india/state"
map_state_tran_list = os.listdir(map_tran_path)
# map_state_tran_list


Map_trans = {'State': [], 'Year': [], 'Quater': [], 'District': [], 'count': [], 'amount': []}

for i in map_state_tran_list:
    p_i = os.path.join(map_tran_path, i)  # Use os.path.join to construct paths
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Use os.path.join to construct paths
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Use os.path.join to construct paths

            try:
                with open(p_k, 'r') as Data:
                    C = json.load(Data)

                    for x in C["data"]["hoverDataList"]:
                        District = x["name"]
                        count = x["metric"][0]["count"]
                        amount = x["metric"][0]["amount"]

                        Map_trans["District"].append(District)
                        Map_trans["count"].append(count)
                        Map_trans["amount"].append(amount)
                        Map_trans['State'].append(i)
                        Map_trans['Year'].append(j)
                        Map_trans['Quater'].append(int(k.strip('.json')))
            except:
                continue

# Create a DataFrame
map_trans_df = pd.DataFrame(Map_trans)


# Specify the directory path where you want to save the CSV file
output_dir = "extracted_data"

# File path for saving the CSV file inside the "data" folder
output_file_path = os.path.join(output_dir, "map_transaction.csv")

# Save the DataFrame to the specified file path
map_trans_df.to_csv(output_file_path, index=False)

# Display the DataFrame
map_trans_df
```

### MAP USER

```python
map_user_path = "pulse/data/map/user/hover/country/india/state"
map_state_user_list = os.listdir(map_user_path)
# map_state_user_list

map_user = {"State": [], "Year": [], "Quater": [], "District": [], "RegisteredUser": []}

for i in map_state_user_list :
    p_i = os.path.join(map_user_path, i)  # Use os.path.join to construct paths
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Use os.path.join to construct paths
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Use os.path.join to construct paths

            try:
                with open(p_k, 'r') as Data:
                    D = json.load(Data)

                    for u in D["data"]["hoverData"].items():
                        district = u[0]
                        registereduser = u[1]["registeredUsers"]

                        map_user["District"].append(district)
                        map_user["RegisteredUser"].append(registereduser)
                        map_user['State'].append(i)
                        map_user['Year'].append(j)
                        map_user['Quater'].append(int(k.strip('.json')))
            except:
                continue

# Create a DataFrame
map_user_df = pd.DataFrame(map_user)

# Specify the directory path where you want to save the CSV file
output_dir = "extracted_data"

# File path for saving the CSV file inside the "data" folder
output_file_path = os.path.join(output_dir, "map_user.csv")

# Save the DataFrame to the specified file path
map_user_df.to_csv(output_file_path, index=False)

# Display the DataFrame
map_user_df
```

## TOP

### TOP TRANSCTION

```python
top_tran_path = "pulse/data/top/transaction/country/india/state"
top_state_tran_list=os.listdir(top_tran_path)
# top_state_tran_list


top_trans = {'State': [], 'Year': [], 'Quater': [], 'District': [], 'Trans_count': [],
             'Trans_amount': []}

for i in top_state_tran_list:
    p_i = os.path.join(top_tran_path , i)  # Use os.path.join to construct paths
    if not os.path.exists(p_i):
        print(f"Directory not found: {p_i}")
        continue

    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Use os.path.join to construct paths
        if not os.path.exists(p_j):
            print(f"Directory not found: {p_j}")
            continue

        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Use os.path.join to construct paths
            if not os.path.exists(p_k):
                print(f"File not found: {p_k}")
                continue

            with open(p_k, 'r') as Data:
                E = json.load(Data)

                for z in E['data']['pincodes']:
                    Name = z['entityName']
                    count = z['metric']['count']
                    amount = z['metric']['amount']
                    top_trans['District'].append(Name)
                    top_trans['Trans_count'].append(count)
                    top_trans['Trans_amount'].append(amount)
                    top_trans['State'].append(i)
                    top_trans['Year'].append(j)
                    top_trans['Quater'].append(int(k.strip('.json')))

# Create a DataFrame
top_trans_df = pd.DataFrame(top_trans)

# Specify the directory path where you want to save the CSV file
output_dir = "extracted_data"

# File path for saving the CSV file inside the "data" folder
output_file_path = os.path.join(output_dir, "top_transaction.csv")

# Save the DataFrame to the specified file path
top_trans_df.to_csv(output_file_path, index=False)

# Display the DataFrame
top_trans_df
```

### TOP USER

```python
top_user_path = "pulse/data/top/user/country/india/state"
top_state_user_list=os.listdir(top_user_path)
# top_state_user_list

top_user = {'State': [], 'Year': [], 'Quarter': [], 'District': [],
            'RegisteredUser': []}

for i in top_state_user_list:
    p_i = os.path.join(top_user_path, i)  # Use os.path.join to construct paths
    if not os.path.exists(p_i):
        print(f"Directory not found: {p_i}")
        continue

    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Use os.path.join to construct paths
        if not os.path.exists(p_j):
            print(f"Directory not found: {p_j}")
            continue

        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Use os.path.join to construct paths
            if not os.path.exists(p_k):
                print(f"File not found: {p_k}")
                continue

            with open(p_k, 'r') as Data:
                F = json.load(Data)

                for t in F['data']['pincodes']:
                    Name = t['name']
                    registeredUser = t['registeredUsers']
                    top_user['District'].append(Name)
                    top_user['RegisteredUser'].append(registeredUser)
                    top_user['State'].append(i)
                    top_user['Year'].append(j)
                    top_user['Quarter'].append(int(k.strip('.json')))

# Create a DataFrame
top_user_df = pd.DataFrame(top_user)

# Specify the directory path where you want to save the CSV file
output_dir = "extracted_data"

# File path for saving the CSV file inside the "data" folder
output_file_path = os.path.join(output_dir, "top_user.csv")

# Save the DataFrame to the specified file path
top_user_df.to_csv(output_file_path, index=False)

# Display the DataFrame
top_user_df

```
